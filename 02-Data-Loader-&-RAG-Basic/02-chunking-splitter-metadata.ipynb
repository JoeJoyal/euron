```python
from langchain_community.document_loaders import TextLoader
```

```python
documents = TextLoader("data.txt").load()
```

```python
documents
```

```python
documents[0].page_content
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=30, chunk_overlap=20, separators="")
splitter.split_documents(documents)
```

Note: ["\n\n", "\n", " ", ""]

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
char_sliter = CharacterTextSplitter(chunk_size = 50, chunk_overlap=10, separator=" ")
```

```python
char_sliter.split_documents(documents)
```

```python
from langchain.text_splitter import TokenTextSplitter
```
