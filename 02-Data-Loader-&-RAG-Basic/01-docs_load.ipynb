```python
import json
from langchain.docstore.document import Document

def load_notion_json(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    documents = []
    for entry in data:
        page_content = f"{entry['records']}\r\n_id: {entry['chunk_text']}"
        metadata = {
            "title": entry.get("records", ""),
            "source": path
        }
        documents.append(Document(page_content=page_content, metadata=metadata))
    return documents
```

```python
load_notion_json("notion.json")
```
