```python
import requests
import numpy as np

def generate_embeddings(text):
    url = "https://api.euron.one/api/v1/euri/embeddings"
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer euri-698e554ded59d10cc4bec9b2c00332697486fe48efa31bb0745b24c5865ce945"
    }
    payload = {
        "input": text,
        "model": "text-embedding-3-small"
    }

    response = requests.post(url, headers=headers, json=payload)
    data = response.json()
    
    embedding = np.array(data['data'][0]['embedding'])
    
    return embedding
```

```python
schema = {
    "class": "ageye",
    "vectorizer": "none", # custom embedding used on my case. If you default used with weaviate embedding model abviously charged.
    "properties": [
        {"name": "text", "dataType": ["text"]}
    ]
}
client.schema.create_class(schema)
print("Schema created!")
```

```python
text = "Welcome to ageye! Indoor farming for next generation AI"
```

```python
embeddings = generate_embeddings(text).tolist()
```
