```python
import chromadb
from chromadb.utils import embedding_functions
```

```python
import requests
import numpy as np

def generate_embeddings(text):
    url = "https://api.euron.one/api/v1/euri/embeddings"
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer euri-698e554ded59d10cc4bec9b2c00332697486fe48efa31bb0745b24c5865ce945"
    }
    payload = {
        "input": text,
        "model": "text-embedding-3-small"
    }

    response = requests.post(url, headers=headers, json=payload)
    data = response.json()
    
    embedding = np.array(data['data'][0]['embedding'])
    
    return embedding
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection(name='ageye-metadata')
```

```python
embeddings_all = [generate_embeddings(i).tolist() for i in dataset]
```

```python
collection.add(
    documents=dataset,
    embeddings=embeddings_all,
    ids=[f"rec_{i}" for i in range(len(dataset))]
)
```

```python
query_emb = generate_embeddings(query).tolist()
```
